```python
import numpy as np
import matplotlib.pyplot as plt

# Sistema de colores jerárquico
hierarchy_colors = {
    'primary': '#1f77b4',      # Insights principales
    'secondary': '#7f7f7f',    # Información de contexto
    'accent': '#ff7f0e',       # Elementos de énfasis
    'warning': '#d62728',      # Problemas/alertas
    'success': '#2ca02c'       # Éxitos/positivos
}

# Aplicar jerarquía en visualizaciones
def apply_hierarchy(ax, data, hierarchy_level='primary'):
    color = hierarchy_colors[hierarchy_level]
    linewidth = 3 if hierarchy_level == 'primary' else 2 if hierarchy_level == 'secondary' else 1
    alpha = 1.0 if hierarchy_level == 'primary' else 0.7 if hierarchy_level == 'secondary' else 0.5

    return ax.plot(data['x'], data['y'], color=color, linewidth=linewidth, alpha=alpha)

# -------- EJEMPLO REAL --------
x = np.linspace(0, 10, 100)

data1 = {'x': x, 'y': np.sin(x)}
data2 = {'x': x, 'y': np.cos(x)}

fig, ax = plt.subplots(figsize=(10, 5))

apply_hierarchy(ax, data1, 'primary')
apply_hierarchy(ax, data2, 'accent')

ax.set_title("Ejemplo usando sistema de colores jerárquico")
ax.grid(True, alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------
# Datos de ejemplo
# -------------------------
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Un punto clave para enfatizar
highlight = {'x': 5, 'y': np.sin(5)}

# -------------------------
# Jerarquía de colores
# -------------------------
hierarchy_colors = {
    'primary': '#1f77b4',
    'secondary': '#7f7f7f',
    'accent': '#ff7f0e',
    'warning': '#d62728',
    'success': '#2ca02c'
}

def apply_hierarchy(ax, x, y, hierarchy_level='primary'):
    color = hierarchy_colors[hierarchy_level]
    lw = 3 if hierarchy_level == 'primary' else 2
    ax.plot(x, y, color=color, linewidth=lw)

# -------------------------
# Flecha de atención
# -------------------------
def add_attention_arrow(ax, start_point, end_point, text='', arrowstyle='->'):
    ax.annotate(text, xy=end_point, xytext=start_point,
                arrowprops=dict(arrowstyle=arrowstyle, color='red', linewidth=2),
                fontsize=12, fontweight='bold')

# -------------------------
# Gradiente de atención
# -------------------------
def attention_gradient_plot(ax, x, y):
    alphas = np.linspace(1, 0.2, len(x))
    for i in range(len(x)):
        ax.scatter(x[i], y[i], alpha=alphas[i], s=60, color='purple')

# -------------------------
# Zoom sobre punto importante
# -------------------------
def emphasize_with_zoom(ax, data_point, zoom_factor=0.5):
    circle = plt.Circle((data_point['x'], data_point['y']),
                        radius=zoom_factor, fill=False,
                        color='red', linewidth=2, linestyle='--')
    ax.add_artist(circle)

    ax.annotate('INSIGHT CLAVE',
                xy=(data_point['x'], data_point['y']),
                xytext=(data_point['x']+1, data_point['y']+0.5),
                arrowprops=dict(arrowstyle='->', color='red', linewidth=2),
                fontsize=12, fontweight='bold', color='red')

# -------------------------
# FIGURA FINAL
# -------------------------
fig, ax = plt.subplots(figsize=(12,6))

apply_hierarchy(ax, x, y, hierarchy_level='primary')
attention_gradient_plot(ax, x, y)

add_attention_arrow(ax,
                    start_point=(2, 0.5),
                    end_point=(highlight['x'], highlight['y']),
                    text='Punto importante')

emphasize_with_zoom(ax, highlight)

ax.set_title("Ejemplo con técnicas de comunicación visual", fontsize=16)
ax.grid(True, alpha=0.3)

plt.show()
```

```python
# Técnica de contraste extremo
def extreme_contrast(ax, normal_data, critical_insight):
    # Datos normales en gris sutil
    ax.plot(normal_data['x'], normal_data['y'], color='#cccccc', linewidth=1, alpha=0.5)

    # Insight crítico en rojo brillante
    ax.plot(critical_insight['x'], critical_insight['y'],
            color='#ff0000', linewidth=4, marker='o', markersize=8)

    # Etiqueta de alto impacto
    ax.text(critical_insight['x'][-1], critical_insight['y'][-1] + 10,
            '¡ALERTA CRÍTICA!', fontsize=16, fontweight='bold',
            color='#ff0000', ha='center')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de ejemplo
x = np.linspace(0, 10, 50)
y_normal = np.sin(x) * 10 + 50          # Datos "normales"
y_critical = np.array([y_normal[30]])  # Punto crítico
x_critical = np.array([x[30]])

normal_data = {'x': x, 'y': y_normal}
critical_insight = {'x': x_critical, 'y': y_critical}

# Crear figura y ejes
fig, ax = plt.subplots(figsize=(8,5))

# Aplicar técnica de contraste extremo
extreme_contrast(ax, normal_data, critical_insight)

# Ajustes visuales
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_title('Ejemplo: Contraste Extremo en Insights')

plt.show()
```
